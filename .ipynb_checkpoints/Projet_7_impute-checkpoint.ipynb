```python
import pandas as pd
import numpy as np
import os
import missingno as msno
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
os.chdir('..')
path = os.getcwd()
train_df = pd.read_csv(path+'\\Project_files\\application_train.csv')
```

```python
def preprocessing(df):
    #set the index as the id of the loan
    df.set_index('SK_ID_CURR')
    #wrong gender code
    df = df.drop(df.loc[df['CODE_GENDER'] == 'XNA'].index)
    #check if binary features are binary
    binary = [i for i in train_df.columns if 'FLAG' in i]
    for i in binary:
        if len(df[i].unique()) != 2:
            print('Column {} is not binary.'.format(i))
    #OHE categorical features
    cat_df = df.select_dtypes(include = ['object'])
    df = df.select_dtypes(exclude = ['object'])
    ohe = OneHotEncoder(sparse = False)
    cat_df = ohe.fit_transform(cat_df)
    colnames = ohe.get_feature_names_out(ohe.feature_names_in_)
    cat_df = pd.DataFrame(cat_df, columns = colnames)
    df = df.join(cat_df)
    #extract target
    if df['TARGET'].isna().sum() != 0:
        df = df.drop(df.loc[df['TARGET'].isna()])
    y = df['TARGET']
    df = df.drop(labels = 'TARGET', axis = 1)
    colnames = df.columns
    return df, y, colnames 
```

```python
df, y, colnames = preprocessing(train_df)
```

```python
num_df = train_df.select_dtypes(exclude = ['object'])
msno.matrix(num_df)
```

```python
imputer = KNNImputer(n_neighbors=5)
impute_df = imputer.fit_transform(num_df) 
```

```python
impute_df = pd.DataFrame(impute_df, columns = num_df.columns)
impute_df.to_csv('impute_df.csv', index = False)
```
